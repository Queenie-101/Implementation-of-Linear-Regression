```python
#2. Library importation
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #splits the data into testing and training
from sklearn.linear_model import LinearRegression #linear regression algorithm
```

```python
#3. Read the data
df=pd.read_csv('petrol_consumption.csv')
df.head()
```

```python
#4. Dataset Information
df.info()
```

```python
#5. Check for null values
df.isnull().sum()
```

```python
from re import X
#6. Split the data into target features
y=df['Petrol_Consumption']
X=df[['Average_income','Paved_Highways','Population_Driver_licence(%)']]
```

```python
#7. Split teh data into testing and training
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
#8. Load the algorithm
algorithm=LinearRegression()
```

```python
#9. Train the algorithm with our data
algorithm.fit(X_train,y_train)
```

```python
#10. Maken predictions
predictions=algorithm.predict(X_test)
```

```python
#11. Compare the predicted values with the actual value
pd.DataFrame({'Actual':y_test,'Predicted':predictions})
```

```python
#12. Evaluate the accuarcy of the algorith.
algorithm.score(X_test,y_test)
```
